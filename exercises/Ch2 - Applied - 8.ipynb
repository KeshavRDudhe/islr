### (a)

Learn to read data from a csv file

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
sns.set(style="darkgrid", color_codes=True)
```

```python
college = pd.read_csv('../data/College.csv')
college
```

### (b) 

Modify columns and set a new index column

```python
cols = list(college.columns)
cols[0] = 'College'
college.columns = cols
college.set_index('College', inplace=True)
college.head()
```

### (c)

#### (i)

Use DataFrame.describe() to produce a numerical summary of the variables

```python
college.describe()
```

#### (ii)

#### (iii)

Produce boxplot for **Outstate** vs **Private**

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Private", y="Outstate", data=college)
```

#### (v)

Produce histograms on quantitative variables

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
sns.distplot(college['Top10perc'], ax=ax1, bins=20);
sns.distplot(college['Top25perc'], ax=ax2, bins=20);
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
sns.distplot(college['S.F.Ratio'], ax=ax1, bins=20);
sns.distplot(college['Grad.Rate'], ax=ax2, bins=20);
```

#### (vi)

Explore the data and provide a brief summary.

```python
cm = college.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(cm, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

mask = ~mask
```

```python
fig, ax = plt.subplots(figsize=(16, 16))
sns.heatmap(cm, annot=True, mask=mask, vmin=0, vmax=1, linewidths=1, cmap='YlGnBu', square=True)
```
